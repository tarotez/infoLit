```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
```

```python
# boxplot
sns.set_theme(style="ticks",palette="pastel")
ax = sns.boxplot(x="smoker", y="tip",
            hue="time", palette=["b","y"],
            data=tips)            
ax.set_title("Tip amount", fontsize=16)
sns.despine(offset=10, trim=True)
fig = ax.get_figure()
fig.savefig("tip.png") 
plt.show()
```

```python
# scatter plot visualizing three attributes using color
iris = sns.load_dataset("iris")
# iris = pd.read_csv("iris.csv")  # if you cannot download directly from seaborn.
sns.set_theme(style="ticks", palette="pastel")
ax = sns.pairplot(iris, hue="species")
plt.show()
```

```python
# scatter plot visualizing four attributes using color and size
miles_per_gallon = sns.load_dataset("mpg")
# miles_per_gallon = pd.read_csv("mpg.csv")  # if you cannot download directly from seaborn.
sns.set_theme(style="whitegrid")
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=miles_per_gallon,
    x="mpg", y="acceleration",
    hue="model_year", size="cylinders",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
plt.show()
```

```python
# use a long-form organization to plot by year
flights = sns.load_dataset("flights")
# flights = pd.read_csv("flights.csv")  # if you cannot download directly from seaborn.
ax = sns.relplot(data=flights, x="year", y="passengers", hue="month", kind="line")
plt.show()
```

```python
# use a wide-form organization to plot by months
flights_wide_form = flights.pivot(index="year", columns="month", values="passengers")
sns.relplot(data=flights_wide_form.transpose(), kind="line")
plt.show()
```

```python
# group by year
ax = sns.lineplot(
    data=pd.DataFrame(flights.groupby(["year"])["passengers"].sum()), 
    x="year", y="passengers")
plt.show()
```
